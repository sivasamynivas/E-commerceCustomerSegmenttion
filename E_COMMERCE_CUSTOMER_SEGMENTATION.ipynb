```python
#- E-COMMERCE CUSTOMER SEGMENTATION
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
df = pd.read_excel("/content/cust_data.xlsx")
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Gender'] = df['Gender'].map({"M":1,"F":2})
data = df
columns_for_clustering = ['Cust_ID', 'Gender', 'Orders', 'Jordan', 'Gatorade', 'Samsung', 'Asus', 'Udis', 'Mondelez International', 'Wrangler', 'Vans', 'Fila', 'Brooks', 'H&M', 'Dairy Queen', 'Fendi', 'Hewlett Packard', 'Pladis', 'Asics', 'Siemens', 'J.M. Smucker', 'Pop Chips', 'Juniper', 'Huawei', 'Compaq', 'IBM', 'Burberry', 'Mi', 'LG', 'Dior', 'Scabal', 'Tommy Hilfiger', 'Hollister', 'Forever 21', 'Colavita', 'Microsoft', 'Jiffy mix', 'Kraft']
silhouette_scores = []
k_values = range(2, 11)  # Range of K values to try
data
for k in k_values:  #Run K-means clustering for different K values
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    labels = kmeans.labels_
    score = silhouette_score(data, labels)
    silhouette_scores.append(score)

plt.plot(k_values, silhouette_scores, 'bx-')# Plot the silhouette scores
plt.xlabel('Number of clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.show()

optimal_k = np.argmax(silhouette_scores) + 2# Select the K value with the highest silhouette score as the optimal K
k = optimal_k# Add 2 because of zero-based indexing
kmeans = KMeans(n_clusters=k, random_state=42)# Apply k-means clustering
kmeans.fit(data)
cluster_labels = kmeans.labels_# Retrieve the cluster labels
data['Cluster'] = cluster_labels# Add the cluster labels to the original dataset
print(data['Cluster'].value_counts())# Print the clusters






```
